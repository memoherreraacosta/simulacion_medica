```python
fig, axs = pp.subplots(figsize=(5, 5), sharey=True)

for column in data.columns:
    print("\n\n\n" + column)
    data[column].value_counts().plot(
        kind="bar",
        fontsize=14,
        title = column
    )
    pp.legend(data[column].values)
    pp.show()
    print(data[column].value_counts())
```
