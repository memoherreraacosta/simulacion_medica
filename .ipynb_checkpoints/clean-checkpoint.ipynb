# Encuesta de Simulaci√≥n

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as pp
import seaborn as sns

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

from pprint import pprint
```

```python
data = pd.read_csv('preguntas.csv')
data = data[data.columns[:-1]]
```

```python
norm_d = {
    column[:2]: data[column].value_counts(normalize=True)*100
    for column in data.columns
}
```

```python
norm_d = pd.DataFrame(norm_d)
dat_t = norm_d.transpose()
```

norm_d

```python
f = pp.figure(figsize=(15,8))

pp.title('Encuesta de simulaci√≥n', color='black')
colorcitos = ["lightgrey", "orchid", "salmon", "skyblue", "springgreen"]
dat_t = dat_t.reindex_axis(["Muy de acuerdo", "De acuerdo", "Neutral", "Desacuerdo", "Muy en desacuerdo"], axis = 1)


ax = dat_t.plot(
    kind="barh",
    stacked=True,
    ax=f.gca(),
    color=colorcitos,
    grid=True,
)

ax.set_xlabel("Porcentajes de resultado")
ax.set_ylabel("No. de Pregunta")
ax.legend(loc='center left',bbox_to_anchor=(1.0, .5))
ax.invert_yaxis()
```
