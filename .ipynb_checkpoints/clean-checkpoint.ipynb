# Encuesta de Simulaci√≥n

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as pp

from pprint import pprint

%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('preguntas.csv')
```

```python
norm_d = {
    column[:2]: round(data[column].value_counts(normalize=True)*100, 2)
    for column in data.columns
}
```

```python
dat_t = pd.DataFrame(norm_d).transpose()
```

```python
f = pp.figure(figsize=(15,8))
pp.title('Encuesta de simulaci√≥n', color='black')
colorcitos = [
    "seagreen", # Muy de acuerdo
    "limegreen", # De acuerdo
    "salmon", # Neutral
    "orangered", # Desacuerdo
    "crimson"#"Muy en desacuerdo
]

dat_t = dat_t.reindex_axis([
        "Muy de acuerdo",
        "De acuerdo",
        "Neutral",
        "Desacuerdo",
        "Muy en desacuerdo"
    ], 
    axis = 1
)

ax = dat_t.plot(
    kind="barh",
    stacked=True,
    ax=f.gca(),
    color=colorcitos,
    grid=True,
)

ax.set_xlabel("Porcentajes de resultado")
ax.set_ylabel("No. de Pregunta")
ax.legend(loc='center left',bbox_to_anchor=(1.0, .5))
ax.invert_yaxis()
```
